```r
generateSquares <- function(x) {   
    for(i in 1:x) {
      y <- i^2
      print(y)
   }
    return(x^2)
}
# Calling the function
a = generateSquares(4)
print(a)
```

```r
sales1<-c(12,14,16,29,30,45,19,20,16, 19, 34, 20)
sales2<-rpois(12,34)  # random numbers, Poisson distribution, mean at 34, 12 numbers
par(bg="cornsilk")

plot(sales1, col="blue", type="o", ylim=c(0,100), xlab="Month", ylab="Sales" )
title(main="Sales by Month")

lines(sales2, type="o", pch=22, lty=2, col="red")
grid(nx=NA, ny=NULL)
legend("topright", inset=.05, c("Sales1","Sales2"), fill=c("blue","red"), horiz=TRUE)
# DISCUSSION: Problem 1: 
# The second sales data is indeed random but we can see that the values are bounded 
# with sales not exceeding value of 60 nor decreasing the value of 20. The values 
# of the sales are not extremely large or small.

# Also please find the result images for for diffrent cominations of the "sales 2 data" attached to the 
# .ipynb file in the lab1part1 folder. 
```

```r
# Another example of High resolution world map using ggplot2
library(ggplot2)
map.world <- map_data(map = "world")

# Creating a high resolution World map. 
p1 <- ggplot(map.world, aes(x = long, y = lat, group = group))
p1 <- p1 + geom_polygon() 
p1 <- p1 + labs(title = "World Map with Plain Regions")

# Creating a High resolution World Map with Colored Regions.
p2 <- ggplot(map.world, aes(x = long, y = lat, group = group, fill = region))
p2 <- p2 + geom_polygon() 
p2 <- p2 + theme(legend.position="none") 
p2 <- p2 + labs(title = "World Map with Filled Regions")

# Printing both of the Maps after creation.
p1 
p2
```
