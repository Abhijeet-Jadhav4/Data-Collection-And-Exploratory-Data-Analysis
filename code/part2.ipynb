Part 2: Graph 1 - Influenza Positive Tests Reports to CDC by U.S. Clinical Laboratories, National Summary, 2018 - 2019 Season

```r
library(ggplot2)
data1 <- read.csv(file.choose())    # Use datafile named graph1.csv
week <- data1$Week

totalA <- data1$Total.A
totalB <- data1$Total.B

percPositiveA <- data1$Percent.Positive.A
percPositiveB <- data1$Percent.Positive.B

percPositive <- data1$X..Positive


dataColumn <- c(totalB, totalA)
dataPlot <- c(rep("B",21),rep("A",21))

dataframe <- data.frame(week,dataColumn,percPositive,percPositiveA,percPositiveB)


p <- ggplot(dataframe,aes(x=factor(week),y=dataColumn,fill=dataPlot))+
    geom_bar(stat="identity")+
    scale_x_discrete(breaks=week[c(T,F,T,F)])+
    theme(axis.text.x = element_text(angle = 70, hjust =0.5, vjust = 0.5)) + 
    scale_fill_manual(values = c("#FFCC66","#119900"),name="") + 
    xlab("Week")+
    ylab("Number of Positive Specimens")+
    labs(title="Influenza Positive Tests Reported to CDC by U. S. Clinical Laboratories,
                        National Summary, 2018-2019 Season")+
    scale_y_continuous(sec.axis = sec_axis(~./400))
    
p1 <- p+geom_line(aes(x=as.numeric(factor(week)),y=percPositive*400, color = "1"),data=dataframe,stat="identity")
        #scale_fill_manual(values=c("#FFFFFF"))
p2 <- p1+geom_line(aes(x=as.numeric(factor(week)),y=percPositiveA*400, color="2"),size=0.5,linetype="longdash",data=dataframe,stat="identity")
        #scale_fill_manual(values=c("#FFFF00"))
p3 <- p2+geom_line(aes(x=as.numeric(factor(week)),y=percPositiveB*400, color="3"),size=0.5,linetype="longdash",data=dataframe,stat="identity")+
        #theme(axis.text.x = element_text(angle = 70, hjust =0.5, vjust = 0.5),legend.position="topleft")+
        scale_color_manual(label = c("Percent Positive","% Positive Flu A","% Positive Flu B"),values=c("1"="black","2"="yellow","3"="green"),name="")
        # 
p3
```

Part 2: Graph 4 - Influenza Positive Tests Reported to CDC by U. S. Public Health Laboratories, National Summary, 2018-2019 Season

```r
library(ggplot2)
data <- read.csv(file.choose())    # Use datafile named graph2.csv

dataColumn <- c(data$A.H3N2v., data$A.H1N1.pdm09, data$A.H3., data$A.Subtyping.not.performed., data$B, data$BVIC, data$BYAM)

dataPlot <- c(rep("4 H3N2v",21),rep("2 A (H1N1)pdm09",21),rep("3 A (H3N2)",21),
              rep("1 A (Subtyping not performed)",21),rep("5 B (lineage not performed)",21),
              rep("6 B (Victoria Lineage)",21),rep("7 B (Yamagata Lineage)",21))

g2dataframe <- data.frame(Week1 = data$Week, dataColumn)

Week <- factor(g2dataframe$Week1)
Number.of.Positive.Specimens <- dataColumn

cols <- c("#ECDB54", "#FFD662","#E94B3C", "#663399","#79C753", "#ADFF2F", "#BFD641")

p <- ggplot(g2dataframe,aes(x=Week, y = Number.of.Positive.Specimens,fill=dataPlot)) + 
        geom_bar(stat="identity") + 
        scale_x_discrete(breaks=Week[c(T,F,T,F)])+
        theme(axis.text.x = element_text(angle = 70, hjust =0.5, vjust = 0.5)) + 
        scale_fill_manual(labels = c("A (Subtyping not performed)","A (H1N1)pdm09","A (H3N2)","H3N2v","B (lineage not performed)","B (Victoria Lineage)","B (Yamagata Lineage)"), values = cols,name="") + 
        xlab("Week") + 
        ylab("Number of Positive Specimens") + 
        labs(title="Influenza Positive Tests Reported to CDC by U. S. Public Health Laboratories,
                        National Summary, 2018-2019 Season")
p
```

Part 2: Graph 6 - Pneumonia and Influenza Mortality from the National Center for Health Statistics Mortality Surveillance System

```r
mortality <-read.csv(file.choose())

mortality$Year.Week <- paste(mortality$Year,"(" ,mortality$Week, ")")
mortality$Year.Week<-gsub(" ","",mortality$Year.Week)
#mortality
x <- as.character(mortality$Year.Week)
x <- factor(x, levels=unique(x))
plot(x, mortality$Expected,main = "Pneumonia and Influenza Mortality from the 
National Center for Health Statistics Mortality Surveillance System", xlim = c(269,490),ylim=c(4,12), 
     xlab="MMWR Week", ylab="% of All Deaths Due to P&I", cex.axis= 0.5)
lines(x, mortality$Threshold,col="black", lwd=2)
lines(x, mortality$Percent.of.Deaths.Due.to.Pneumonia.and.Influenza, col="red", lwd=2)
lines(x, mortality$Expected,col="black", lwd=2)


legend("bottom",
       270,0, 
       legend = c("Epidemic Threshold", "Seasonal Baseline"), 
       col = c("red", "black"),
       lty=c(1,1),
       #lwd=2,
       #bty = "n",
       #text.col = "black", 
       horiz = T  )
    #inset = c(0.1, 0.1))
```

Part 2: Graph 7 - Number of Influenza-Associated Pediatric Deaths by Week of Death: 2015-2016 season to present

```r
library(ggplot2)
data1 <- read.csv(file.choose())
dataColumn <- c(data1$PREVIOUS.WEEK.DEATHS, data1$CURRENT.WEEK.DEATHS)
dataPlot <- c(rep("PREVIOUS WEEK DEATHS",208),rep("CURRENT WEEK DEATHS",208))
dataframe <- data.frame(Week1 = data1$WEEK.NUMBER, dataColumn)
Week <- factor(dataframe$Week1)
p <- ggplot(dataframe,aes(x=Week, y = dataColumn,fill=dataPlot)) + 
        geom_bar(stat="identity") + 
        xlab("Week of Death") +  
        ylab("Number of Deaths") +
        scale_x_discrete(breaks=data1$WEEK.NUMBER[c(T,F,F,F,F,F,T,F,F,F,F,F)])+
        theme(plot.title=element_text(hjust=0.35), axis.text.x=element_text(angle=90, hjust=1,vjust=0.5),
        legend.position="bottom", legend.box.background=element_rect(colour="black",size=1))+
        labs(title="      Number of Influenza-Associated Pediatric Deaths \n   by Week of Death: 2015-2016 season to present")+
        annotate("text",x="201552",y=22.5,label="      2015-2016")+
        annotate("text",x="201652",y=22.5,label="2016-2017")+
        annotate("text",x="201748",y=22.5,label="2017-2018")+
        annotate("text",x="201848",y=22.5,label="2018-2019")+
        annotate("text",x="201549", y=21, label="           Number of Deaths\n         Reported = 94",size=3)+
        annotate("text",x="201652", y=21, label="Number of Deaths\n Reported = 110",size=3)+
        annotate("text",x="201748", y=21, label="Number of Deaths\n Reported = 185",size=3)+
        annotate("text",x="201848", y=21, label="Number of Deaths\n Reported = 85",size=3)+
        scale_fill_manual(values=c("cyan","dark green"),name="")
        
p 
```
